```python
import csv

m = []
for i in range(1000):
    m.append(i)
    
with open('blabla.csv', mode='w') as employee_file:
    employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    employee_writer.writerow(m)

```
